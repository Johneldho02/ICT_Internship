## Extracting only the numerical columns for clustering

```python
num_tr=tr.select_dtypes(include=['float64','int64'])
num_tr
```

## Scaling the numerical columns between the values 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()

mm_tr=mm.fit_transform(num_tr)
scaled=pd.DataFrame(mm_tr)
scaled.columns=num_tr.columns
scaled
```

## Principal Component Analysis (to make the clustering more efficient)

```python
from sklearn.decomposition import PCA 
pca=PCA().fit(scaled)

plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 8, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 8, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.97, color='r', linestyle='-')
plt.text(0.5, 0.85, '97% cut-off threshold', color = 'red', fontsize=16)
# The above line of code, draws a line at 0.97
ax.grid(axis='x')
plt.show()
```

### Since the no. of components is 2, under the threshold, we choose the no. of components as 2

```python
pca=PCA(n_components=2)
pca.fit(scaled)
trr=pca.transform(scaled)
TR=pd.DataFrame(trr)
TR
```

#### Extracting only the values without labels

```python
t=TR.values
```

# DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN   
from sklearn.metrics import silhouette_score as ss

db=DBSCAN(eps=0.5357142857142857, min_samples=2).fit(t)
print('Silhouette Score is ',ss(t,db.labels_))
```

## Checking the different labels of the clusters 

```python
clu=pd.DataFrame(db.labels_)
clu.value_counts()
```

### Adding the cluster labels into the dataset as a new column

```python
ttt=TR.copy()
ttt['clusters']=db.labels_
ttt.clusters.value_counts().to_frame()
```

## Plotting of the clusters
### The diagram shows well separated clusters.

```python
outliers = ttt[ttt['clusters']==-1]

fig2, (axes) = plt.subplots(1,figsize=(12,5))
sns.scatterplot(x=0,y=1,

                data=ttt[ttt['clusters']!=-1],

                hue='clusters', palette='Set2', legend='full', s=200)


axes.scatter(outliers[0], outliers[1], s=10, label='outliers', c="k")

axes.legend()

plt.setp(axes.get_legend().get_texts(), fontsize='12')

plt.show()
```
